# **4. Instalando o Portainer**

 Vamos instalar o portainer pra facilitar o gerenciamento dos containers, já que é uma interface gráfica.

---

**4.1** Vamos criar um volume para dados do portainer.

```python
#docker volume create portainer_data
```

**4.2** Vamos executar o container do portainer

```python
'''
    docker run -d -p 8000:8000 -p 9443:9443 \
    --name=portainer \
    --restart=always \
    -v /var/run/docker.sock:/var/run/docker.sock \
    -v portainer_data:/data \
    portainer/portainer-ce:latest
'''
```

**4.3** Agora vamos acessar o portainer

---
